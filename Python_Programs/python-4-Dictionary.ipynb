```python
# sample keys using values
student = {"id": 101, "name": "Alice", "age":20, "grade": "A"}
```

```python
# valyes to serch for 
search_value = 20
```

```python
# method 1: using a loop
keys_with_value = []
for key,value in student.items():
    if value == search_value:
        keys_with_value.append(key)

print("keys for value", search_value, ":", keys_with_value)
# Output :keys for value 20
    
```

```python
# method 2: Using list comperhension (shorter  way)
keys_with_value2 = [key for key, value in student.items() if value == search_value]
print("keys found (list comprehension):", keys_with_value2)
```
