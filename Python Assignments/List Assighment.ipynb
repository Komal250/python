```python
# 13. Recreate a list and check whether a specific element (for example "apple") exists in the list using the 'in' operator.
fruits = ["apple", "banana", "cherry", "mango", "banana"]
print(fruits)
print("Is 'apple' in fruits?", "apple" in fruits)
```
