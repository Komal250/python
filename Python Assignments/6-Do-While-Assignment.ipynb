```python
# 15Write a program that takes a number N as input and prints the multiplication table of N from 1 to 10 using a while loop.
num = int(input("Enter number: "))
b = 1

while b <= 10:
    print(num, "a", b, "=", num * b)
    b += 1
```

```python
# 16Write a program that takes a number as input and calculates the sum of its digits using a while loop.
number = int(input("Enter a number: "))
total = 0

while number > 0:
    digit = number % 10
    total += digit
    number //= 10

print("Sum of digits:", total)
```

```python
# 17. Write a program that takes a number N as input and prints all numbers from N down to 1 in reverse order using a while loop.
n = int(input("Enter  number: "))
while n >= 1:
    print(n)
    n -= 1
```

```python
# 18Write a program that takes a number as input and finds its factorial using a while loop.
number = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= number:
    factorial *= i
    i += 1
print("Factorial:", factorial)
```
