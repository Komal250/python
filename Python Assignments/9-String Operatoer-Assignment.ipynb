```python
# 27 Write a program to count uppercase and lowercase letters in a string.
string = input("enter your string:") 
upper_case = 0
lower_case = 0
for char in string:
    if char.isupper():
        upper_case +=1
    elif char.islower():
        lower_case +=1
print("num of uppercase letter:", upper_case)
print("num of lower letter:", lower_case)
```

```python
# 28 Write a program to remove all special characters from a string.
string = input("enter string")
text = ""
for ch in string:
    if ch.isalnum():
        text +=ch
print("string after removing special char:", text)
```
