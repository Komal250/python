```python
# 21. Find all keys that have a specific value in a dictionary.
# Find all keys that have a specific value in a dictionary

my_dict = {
    "a": 10,
    "b": 20,
    "c": 10,
    "d": 30,
    "e": 10
}
target_value = 10
keys = [key for key, value in my_dict.items() if value == target_value]
print("Keys with value", target_value, ":", keys)
```

```python
# 22 Count how many students scored above 80 marks
marks = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 90,
    "David": 67,
    "Eva": 88
}
count = sum(1 for score in marks.values() if score > 80)
print("Number of students scoring above 80:", count)
```

```python
# 25. Update the department of a student in the nested dictionary.
# Update the department of a student in a nested dictionary

students = {
    "101": {"name": "Alice", "department": "Computer Science"},
    "102": {"name": "Bob", "department": "Mechanical"},
    "103": {"name": "Charlie", "department": "Electrical"}
}
students["102"]["department"] = "Civil"
print(students)
```

```python
# 26  Loop through the nested dictionary and print student names with their grades.
# Loop through the nested dictionary and print student names with their grades

students = {
    "101": {"name": "Alice", "grade": "A"},
    "102": {"name": "Bob", "grade": "B"},
    "103": {"name": "Charlie", "grade": "A"}
}

for student_id, details in students.items():
    print(details["name"], ":", details["grade"])
```

```python
# 27 Merge two dictionaries of students into one

students1 = {
    "101": "Alice",
    "102": "Bob"
}
students2 = {
    "103": "Charlie",
    "104": "David"
}
merged_students = {**students1, **students2}
print("Merged Dictionary:")
print(merged_students)
```
