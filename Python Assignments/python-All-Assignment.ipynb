```python
# 2. Use while loop to calculate the factorial of a number without using math module.
number = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= number:
    factorial *= i                      # fact means 1x2x3 like
    i += 1                              # fact *i ne multiplicatin hot up to loop
print("Factorial:", factorial)
```

```python
# 3. Find the sum of all digits of a number using while loop.
number = int(input("enetr your number:",))
sum = 0 
while number >0:
    digit = number % 10          # % means last digit
    sum = sum + digit
print("sum of  digit :", sum)
```

```python
# 4. Print the reverse of a string using while loop.
string = input("enter your string :")
i = len(string - 1)
while i >=0:
    print(string[i], end="")           # len string -1 means last index 
    i -= 1                             # i -= 1 ne index kami keryto
```

```python
# 5. Write a while loop to count how many times a character appears in a string.
text = input("Enter a string: ")
char = input("Enter a character to search: ")

count = 0
i = 0
while i < len(text):
    if text[i] == char:
        count += 1
    i += 1
print("The character", char, "appears", count, "times.")
```

```python
# 7. Use nested for loops to print a hollow square pattern.
# Output:
# *****
# *   *
# *   *
# *   *
# *****
num = 5                                       # 5 rows ,first row last row ,first colum last column = * innner space
for i in range(num):                          # i=0 first row star .row 2,3,4  (1= star, sapce last star)
    if i == 0 or i == num-1:                  #  row i=4 last row star complete.
        print("*" * num)                      # i == 0 first row
    else:                                     # i == num-1 last row
        print("*" + " "*(num-2) + "*")        # space complte
```

```python
# 8. Write a for loop to print a number pyramid.
# Output:
#      1
#    121
#   12321
#  1234321
rows = 4
for i in range(1, rows + 1):        #bahercha loop and # i = row number    
    print(" " * (rows - i), end="") # row-1 mhanje i space , 1=3 cha space 2=2 ,3=1 cha space 4=0 var jast space khali kami 
    for j in range(1, i + 1):       # 1 pasun i prnt num printed
        print(j, end="")            
    for j in range(i - 1, 0, -1):   # i -1 revers print karto 
        print(j, end="")            # end="" mhanje new line

    print()
```
