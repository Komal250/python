```python
# function with parametrs (inpute)

# function defination
def greet_user(name):
    print("Hello", name, "I Welcome to python Functions")

# Functions call
greet_user("Rahul")
greet_user("Priya")
```

```python
# function with mutiplal statement
def calculate_square_and_cube(number):
    square = number **2 
    cube = number ** 3
    return square, cube # retrun multiple value 

# function call
sq, cb = calculate_square_and_cube(4)
print("Square:", sq)
print("Cube:", cb)
```

```python
# Function calling another function 
def add(a,b):
    return a+b

def multiply(a,b):
    return a*b
def opration(x,y):
    total = add(x,y)
    product = multiply(x,y)
    print("Addition:", total)
    print("Multiplication:", product)

# Functin call
opration(5,3)
```

```python
# Function with condinal logic inside

def check_even_odd(number):
    if number % 2==0:
         return "Even"
    else:
        return "Odd"

print("5 is", check_even_odd(5))
print("10 is", check_even_odd(10))
```
