```python
# Enclosing and Nonlocal variable (nested functions)
```

```python
def outer_function():
    value = 10 # Enclosing variables 

    def inner_function():
        nonlocal value # refer to variable from the outer function 
        value += 5 
        print("Inside Inner Function :", value)

    inner_function()
    inner_function()

outer_function()
```
