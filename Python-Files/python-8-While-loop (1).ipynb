```python
# Example : print number from 1 to 10 using while loop

# step 1: Initialize the starting number 
num = 1 

# step 2: Start the while loop 
# The loop will run as long as 'num' is less than or equl to 10 
while num <= 10:
    print(num)     # step 3: print the current value of 'num'
    num = num + 1  # step 4: Incress 'num' by 1 in each interation
```

```python
# step 1: Initialize the starting number 
num = 50
while num >= 1:
    print(num)    
    num = num -1
```

```python
count = 1 
while count <= 5:
    print("count:", count)
    count += 1
```

```python
count = 10 
while count >= 1:
    print("count:", count)
    count -= 1
```

```python
# printing odd numbers from 1 to 10 using while loop 

num = 1 
while num <= 10:
    if num % 2!= 0:
        print(num)
    num = num + 1
```

```python
# printing even numbers from 50 to 1 using while loop

num = 50
while num <= 1:
    print(num)
num = num -=1 
```

```python
num = 1 
while num <= 10:
    if num % 2 == 0:
       print(num, "is Even")
    else:
        print(num, "is Odd")
    num = num + 1
```

```python
# printing the square of numbers from 1 to 10 using while loop

num = 1 

while num <= 10:
    print("Square of", num, "=", num * num)
    num = num + 1
```

```python
num = 1 
while num <= 50:
    print("Number:", num)
    if num % 2 == 0:
        print("Square:", num * num)
        num += 1
```

```python
# Practical exmaple : Login athhempt system using while lopp 

correct_password = "pythom123"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    user_input = input("Enter your password: ")

    if user_input == correct_password:
        print(" Access Granted! Welcome! ")
    else:
        attempts += 1
        print(" Incorrect password. attempts left:", max_attempts - attempts)

if attempts == max_attempts:
    print(" Access Denied! Too many failed attempts.") 
    
```
