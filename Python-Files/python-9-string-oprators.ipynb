```python
# String Formatting Opratoers 
name = "Seva"
age = 25 
# Old=style Formatting 
print("Using %% Opratoer: Nmane: %s, Age: %d" % (name, age))
# format() method 
print ("Using format(): Name: {}, Age:{}".format(age, name))
# f-string (modern and efficient)
print(f"Using f-string: Name: {name}, Age: {age}")
```
