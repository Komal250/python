```python
# define a parent class called animal
class Animal:
    def sound(self):
        # a generic mathod to be overidden in child classes
        return "Some generic sound"

# Child class 1
class Dog(Animal ):
    def sound(self):
        # Specific inplimentation for dog 
        return "Bark!"

# Child class 2
class Cat(Animal ):
    def sound(self):
        # Specific inplimentation for cat
        return "Meow!"

# Child class 3
class Cow(Animal ):
    def sound(self):
        # Specific inplimentation for cow
        return "Moo!"


animals = [Dog(), Cat(), Cow()]

# Loop through the list call the same method for each object 
for animals in animals:
    print(f"{animals.__class__.__name__} makes sound: { animals.sound()}")


# A function that accepts any objaect with a sound methd
def make_animal_sound(animal):
    print(animal.sound())

# passing different objects to same function
make_animal_sound(Dog())
make_animal_sound(Cat()) 
make_animal_sound(Cow())
```

```python
# Parent Class (Base Class)
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def start_engine(self):
        print(f"{self.brand} {self.model}'s engine started.")
    
    def stop_engine(self):
        print(f"{self.brand} {self.model}'s engine stopped.")


# Child Class (Derived Class) inheriting from Vehicle

class Car(Vehicle):
    def __init__(self, brand, model, fuel_type):
        # super() is used to call the parent class constructor
        super().__init__(brand, model)
        self.fuel_type = fuel_type
    
    def display_details(self):
        print(f"Car Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Fuel Type: {self.fuel_type}")


# Another Child Class inheriting from Vehicle

class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        super().__init__(brand, model)
        self.cc = cc  # engine capacity in cubic centimeters

    def display_details(self):
        print(f"Bike Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Engine Capacity: {self.cc}cc")



# Creating object of Car class
car1 = Car("Toyota", "Fortuner", "Diesel")
car1.start_engine()        # Method inherited from Vehicle
car1.display_details()     # Method from Car class
car1.stop_engine()         # Method inherited from Vehicle

print("\n-----------------------------\n")

# Creating object of Bike class
bike1 = Bike("Yamaha", "R15", 155)
bike1.start_engine()       # Method inherited from Vehicle
bike1.display_details()    # Method from Bike class
bike1.stop_engine()        # Method inherited from Vehicle
```
