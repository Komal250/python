```python
# que 12. wriet a function simple_interest(principal , rate,time) that retrun both intersert total amount

def simple_interset(principal,rate, time):
    interset =(principal* rate* time)/ 100
    total = principal + interset
    return interset, total

# function call 
si,total_amount = simple_interset(1000,5,2)
print("Simple Interset:", si)
print("Total Amount:", total_amount)
```

```python
# que 13 write function calculate_fine(days, fine_per_days=2) that calculates the total fins based on overdue days
# if days <= 0 return "no fine"

def calculate_fine(days, fine_per_day=2):
    if days <= 0:
        return "No Fine"
    else:
        return days * fine_per_day

# function call 
print("fine for  5 days:", calculate_fine(5))
print("fine for 0  days:", calculate_fine(0))
print("fine for 3 days:", calculate_fine(3,3))
```

```python
# 19) Write a function that counts how many vowels are present in a given string.
def count_vowels(text):
    count = 0
    vowels = "aeiouAEIOU"
    for char in text:
        if char in vowels:
            count += 1
        return count 

print(count_vowels(" Komal Shinde"))
```
